```python
'''
向服务器发送get 请求时，分析对‘ie=关键字’中关键字进行parse.urlencode编码，组成完整的fulurl

'''
from urllib import request,parse
baseurl = 'http://www.baidu.com/s?'
ie = input("input your keyword:")
data = {
    'ie':ie
    
}

data = parse.urlencode(data)
fulurl = baseurl + data

headers = {
    'Content-Length':len(data)
      
}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'

req = request.Request(url=fulurl,headers=headers)
rsp = request.urlopen(req)
html = rsp.read().decode('utf-8')
print(html)
```

```python
'''
向服务器发送post请求时，分析百度翻译网页
'''
from urllib import request,parse
import json
baseurl = 'http://fanyi.baidu.com/sug'
kw = input('input your keyword:')
data = {
    'kw':kw
}
data = parse.urlencode(data).encode('utf-8')
print(type(data))
headers={
    'Content-Length':len(data)
}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'
req = request.Request(url=baseurl,data=data,headers=headers)
rsp = request.urlopen(req)
json_data = rsp.read().decode('utf-8')
json_data = json.loads(json_data)
print(json_data)
```

```python
'''
使用User-Agent伪装，try except异常处理
'''
from urllib import request,error
url = 'http://jyj.xiangyang.gov.cn/zxzx/xxhdt/201808/t20180820_1311267.shtml'
headers = {}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'
try:
    req = request.Request(url=url,headers=headers)
    rsp = request.urlopen(req)
    html = rsp.read().decode('utf-8')
    print(html)
except error.HTTPError as e:
    print(e)
except error.URLError as e:
    print(e)
except Exception as e:
    print(e)
```

```python
from urllib import request,error
url = 'http://jyj.xiangyang.gov.cn'

proxy = {'http':'115.159.31.195:8080'} #  设置代理地址
proxy_handler = request.ProxyHandler(proxy) # 创建处理器
opener = request.build_opener(proxy_handler) # 创建opener
request.install_opener(opener) # 安装opener

try:
    req = request.Request(url=url)
    rsp = request.urlopen(req)
    html = rsp.read().decode('utf-8')
    print(html)
except error.HTTPError as e:
    print(e)
except error.URLError as e:
    print(e)
except Exception as e:
    print(e)
    
```

```python
# 模拟人人网的登录-收集cookie
from urllib import request,parse
from http import cookiejar
# 创建cookiejar实例
cookie = cookiejar.CookieJar()
#生成cookie管理器
cookie_handler = request.HTTPCookieProcessor(cookie)
#生成http请求管理器
http_handler = request.HTTPHandler()
#生成https请求管理器
https_handler = request.HTTPSHandler()
#创建请求管理器
opener = request.build_opener(http_handler,https_handler,cookie_handler)

def login():
    url = 'http://www.renren.com/PLogin.do'
    data = {
        'email':'13119144223',
        'password':'123456'
    }
    data = parse.urlencode(data).encode('utf-8')
    req = request.Request(url,data=data)
    # 使用opener发起请求
    rsp = opener.open(req)
    
def getHome():
    url = 'http://www.renren.com/965187997/profile'
    
    rsp = opener.open(url)
    html = rsp.read().decode('utf-8')
    with open('rsp.html','w') as f:
        f.write(html)
if __name__ == '__main__':
    login()
    getHome()
```

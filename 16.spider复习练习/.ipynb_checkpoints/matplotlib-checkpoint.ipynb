```python
import matplotlib.pyplot as plt
import numpy as np

# 从[-1,1]中等距去50个数作为x的取值
x = np.linspace(-1, 1, 50)
print(x)
y = 2*x + 1
# 第一个是横坐标的值，第二个是纵坐标的值
plt.plot(x, y)
# 必要方法，用于将设置好的figure对象显示出来
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x,y1)


plt.figure(figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='r',linewidth=1,linestyle='--')

plt.xlim(-1,2)  # 设置x轴的上下限
plt.ylim(-2,3)  # 设置y轴的上下限
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1,2,5) # 设置新的轴记号
plt.xticks(new_ticks)
plt.yticks([-2,-1,0,1.5,3],
          ['really bad','bad','normal','good','really good'])   # 用单词代替数字

'''# gca = 'get current axis'
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
'''
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x,y1,label='y=2x+1')
plt.plot(x,y2,label='y=x**2')

plt.xlim(-1,2)
plt.ylim(-2,3)
plt.xlabel('X')
plt.ylabel('Y')

new_ticks = np.linspace(-1,2,5)
plt.xticks(new_ticks)

plt.legend(loc='best') # legend图例 简单的方法

plt.show()
```

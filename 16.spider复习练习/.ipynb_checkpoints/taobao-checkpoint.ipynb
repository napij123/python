```python
'''
淘宝男士短袖一页的相关信息爬取，取得想要类型的信息
selenium-driver-xpath
'''
from selenium import webdriver
driver = webdriver.Chrome()
driver.get('https://www.taobao.com/')
driver.implicitly_wait(5)
driver.find_element_by_id('q').clear()
driver.find_element_by_id('q').send_keys('男士短袖')
driver.find_element_by_class_name('btn-search').click()
#print(driver.page_source)

title=[]
titles = driver.find_elements_by_xpath('//div[@class="row row-2 title"]/a')  
for i in titles:
    title.append(i.text)
#print(title)

price = []
prices = driver.find_elements_by_xpath('//div[@class="row row-1 g-clearfix"]/div/strong')
for i in prices:
    price.append(i.text)
#print(price)
#print(len(price))

pay = []
pays = driver.find_elements_by_xpath('//div[@class="deal-cnt"]')
for i in pays:
    pay.append(i.text)
#print(pay)
#print(len(pay))

store = []
stores = driver.find_elements_by_xpath('//div[@class="shop"]/a/span[2]')
for i in stores:
    store.append(i.text)
#print(store)
#print(len(store))

for t,pr,pa,st in zip(title,price,pay,store):
    data={
        'title':t,
        'price':pr,
        'pay_count':pa,
        'store':st
        
    }
    print(data)
```

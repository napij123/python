```python
'''
向服务器发送get 请求时，分析对‘ie=关键字’中关键字进行parse.urlencode编码，组成完整的fulurl

'''
from urllib import request,parse
baseurl = 'http://www.baidu.com/s?'
ie = input("input your keyword:")
data = {
    'ie':ie
    
}

data = parse.urlencode(data)
fulurl = baseurl + data

headers = {
    'Content-Length':len(data)
      
}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'

req = request.Request(url=fulurl,headers=headers)
rsp = request.urlopen(req)
html = rsp.read().decode('utf-8')
print(html)
```

```python
'''
向服务器发送post请求时，分析百度翻译网页
'''
from urllib import request,parse
import json
baseurl = 'http://fanyi.baidu.com/sug'
kw = input('input your keyword:')
data = {
    'kw':kw
}
data = parse.urlencode(data).encode('utf-8')
print(type(data))
headers={
    'Content-Length':len(data)
}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'
req = request.Request(url=baseurl,data=data,headers=headers)
rsp = request.urlopen(req)
json_data = rsp.read().decode('utf-8')
json_data = json.loads(json_data)
print(json_data)
```

```python
'''
使用User-Agent伪装，try except异常处理
'''
from urllib import request,error
url = 'http://jyj.xiangyang.gov.cn/zxzx/xxhdt/201808/t20180820_1311267.shtml'
headers = {}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.2; WOW64; rv:21.0) Gecko/20100101 Firefox/21.0'
try:
    req = request.Request(url=url,headers=headers)
    rsp = request.urlopen(req)
    html = rsp.read().decode('utf-8')
    print(html)
except error.HTTPError as e:
    print(e)
except error.URLError as e:
    print(e)
except Exception as e:
    print(e)
```

```python
# 模拟人人网的登录-收集cookie
from urllib import request,parse
from http import cookiejar
# 创建cookiejar实例
cookie = cookiejar.CookieJar()
#生成cookie管理器
cookie_handler = request.HTTPCookieProcessor(cookie)
#生成http请求管理器
http_handler = request.HTTPHandler()
#生成https请求管理器
https_handler = request.HTTPSHandler()
#创建请求管理器
opener = request.build_opener(http_handler,https_handler,cookie_handler)

def login():
    url = 'http://www.renren.com/PLogin.do'
    data = {
        'email':'13119144223',
        'password':'123456'
    }
    data = parse.urlencode(data).encode('utf-8')
    req = request.Request(url,data=data)
    # 使用opener发起请求
    rsp = opener.open(req)
    
def getHome():
    url = 'http://www.renren.com/965187997/profile'
    
    rsp = opener.open(url)
    html = rsp.read().decode('utf-8')
    with open('rsp.html','w') as f:
        f.write(html)
if __name__ == '__main__':
    login()
    getHome()
```

```python
'''
cookie保存到cookie2.txt文件里,在登录后才有cookie,用cookie.save()
只需要login()函数
'''
from urllib import request,parse
from http import cookiejar
# 创建cookiejar实例
filename = 'cookie2.txt'
cookie = cookiejar.MozillaCookieJar(filename) # 使用FileCookieJar会报错。。
#生成cookie管理器
cookie_handler = request.HTTPCookieProcessor(cookie)
#生成http请求管理器
http_handler = request.HTTPHandler()
#生成https请求管理器
https_handler = request.HTTPSHandler()
#创建请求管理器
opener = request.build_opener(http_handler,https_handler,cookie_handler)

def login():
    url = 'http://www.renren.com/PLogin.do'
    data = {
        'email':'13119144223',
        'password':'123456'
    }
    data = parse.urlencode(data).encode('utf-8')
    req = request.Request(url,data=data)
    # 使用opener发起请求
    rsp = opener.open(req)
    # ignore 忽视意思，就是忽视丢弃的，忽视到期的。都为True
    cookie.save(ignore_discard=True,ignore_expires=True)
if __name__ == '__main__':
    login()
   
```

```python
'''
cookie的读取,在一开始装载已有的cookie文件，用cookie.load()
只需要getHomePage()函数
'''
from urllib import request,parse
from http import cookiejar
cookie = cookiejar.MozillaCookieJar()
cookie.load('cookie2.txt',ignore_discard=True,ignore_expires=True)

cookie_handler = request.HTTPCookieProcessor(cookie)
http_handler = request.HTTPHandler()
https_handler = request.HTTPSHandler()
opener = request.build_opener(http_handler,https_handler,cookie_handler)

def getHomePage():
    url ='http://www.renren.com/965187997/profile'
    
    rsp = opener.open(url)
    html = rsp.read().decode()
    print(html)

getHomePage()
```

```python
'''
requests 下的get请求的方法，参数params,不需要进行parse解码
'''
import requests
url = 'https://www.baidu.com/s?'

data = {
    
    'wd': '时间'
}

headers = {
    
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'
}
rsp = requests.get(url,params=data,headers=headers)
rsp.encoding='utf-8'
#print(rsp.text)
#print(rsp.content) 
print(rsp.status_code) # 状态码
```

```python
'''
百度翻译，requests下的post请求，参数data,不需要解码。
'''
import requests
url = 'http://fanyi.baidu.com/sug'
kw = input('input your word:')
data = {
    
    'kw': kw
}
headers = {
    
    'Content-Length': str(len(data))
        
}
headers['User-Agent']='Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'
rsp = requests.post(url,data=data,headers=headers)
print(rsp.text)
print(rsp.json()) 
```
